```python
covid_df=covid_df.dropna()
```

```python
covid_df.drop_duplicates()
```

```python
covid_df['active_cases']= covid_df['Confirmed']-covid_df['Deaths']+covid_df['Cured']
```

```python
covid_df[['Cured', 'Deaths', 'Confirmed','active_cases']] = covid_df[['Cured', 'Deaths', 'Confirmed','active_cases']].astype(int)
```

```python
covid_df
```

```python
covid_df_copy=covid_df.copy()
```

```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'], axis=1, inplace=True)
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'])
```

```python
covid_df
```

```python
state_wise=pd.pivot_table(covid_df,values= ['Cured','Deaths','Confirmed','active_cases'],index='State/UnionTerritory',aggfunc=max)
state_wise['recovery_rate']=state_wise['Cured']*100/state_wise['Confirmed']
state_wise['death_rate']=state_wise['Deaths']*100/state_wise['Confirmed']
state_wise.sort_values(by='Confirmed',ascending=True)
state_wise.style.background_gradient(cmap='cubehelix')
```

```python
#top 10 active case
top_active_case=covid_df.groupby(by='State/UnionTerritory')['active_cases'].sum().sort_values(ascending=False).reset_index().head(10)
top_active_case
```

```python
##graph for top 10 active case
# Set the style
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x='active_cases', y='State/UnionTerritory', data=top_active_case, palette='viridis')

# Add titles and labels
plt.title('Top 10 States/Union Territories with Highest Active COVID-19 Cases', fontsize=16)
plt.xlabel('Active Cases', fontsize=14)
plt.ylabel('State/Union Territory', fontsize=14)

# Show the plot
plt.show()
```

```python
#top 10 death case
top_death_case=covid_df.groupby(by='State/UnionTerritory')['Deaths'].sum().sort_values(ascending=False).reset_index().head(10)
```

```python
##graph for top 10 active case
# Set the style
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x='Deaths', y='State/UnionTerritory', data=top_death_case, palette='viridis')

# Add titles and labels
plt.title('Top 10 States/Union Territories with Highest COVID-19 Death Cases', fontsize=16)
plt.xlabel('Death Cases', fontsize=14)
plt.ylabel('State/Union Territory', fontsize=14)

# Show the plot
plt.show()
```

```python
##growth trend
fig = plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Andhra Pradesh', 'Tamil Nadu', 'Kerala'])], x='Date', y='active_cases', hue='State/UnionTerritory')
ax.set_title('Growth Trend Of Actiive Cases In India')

plt.show()
```

```python
vacti_df = vacti_df.drop(columns=['Transgender (Doses Administered)',
                                   'Sputnik V (Doses Administered)',
                                   '18-44 Years (Doses Administered)',
                                   '45-60 Years (Doses Administered)',
                                   'AEFI'],
                         axis=1)
```

```python
vacti_df_copy=vacti_df .copy
```

```python
male_doses =vacti_df['Male (Doses Administered)'].sum()
female_doses =vacti_df['Female (Doses Administered)'].sum()
# Data to plot
labels = 'Male', 'Female'
sizes = [male_doses, female_doses]
colors = ['blue', 'pink']
explode = (0.1, 0)  # explode 1st slice (Male)

# Plot the pie chart
plt.figure(figsize=(10, 7))
plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.1f%%', startangle=140)

plt.title('Distribution of Vaccine Doses Administered by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

```python
vaccin=vacti_df[vacti_df.State!='India']
vaccin
```

```python
vaccin.columns
```

```python
state_vaccination_totals =vaccin.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total Individuals Vaccinated')
```

```python
top_5_states = state_vaccination_totals.sort_values(by='Total Individuals Vaccinated', ascending=False).head(5)
top_5_states
```

```python
# Plotting the total doses administered by top 10 states
plt.figure(figsize=(12, 6))

# Select only the 'Total Individuals Vaccinated' column
top_5_states = top_5_states['Total Individuals Vaccinated']

# Convert the column values to numeric
top_5_states = pd.to_numeric(top_5_states)

sns.barplot(x=top_5_states.index, y=top_5_states.values, palette='viridis')
plt.title('Top 5 States by Total Individuals Vaccinated')
plt.xlabel('State')
plt.ylabel('Total Individuals Vaccinated')
plt.xticks(rotation=45)
plt.show()
```
